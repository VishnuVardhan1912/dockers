# Containers Vs Image

A Docker image is an immutable (unchangeable) file that contains the source code, libraries, dependencies, tools, and other files needed for an application to run

A Docker container is a virtualized run-time environment where users can isolate applications from the underlying system.

From Dockerfile to Image to Container.

To build a docker container(use to run applications) we need a docker image.

Docker image is a template to create container. Inside the container we have container layer and multiple image layer. container is in read-write mode and image layer is in read only. 

From Dockerfile(set of instructions) we create image (docker build)from image we run container.(docker create)
